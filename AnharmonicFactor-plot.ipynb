```python
import os

import numpy as np

from interfacePy.AnharmonicFactor import anharmonic_factor


import matplotlib
matplotlib.use('nbAgg')
import matplotlib.pyplot as plt

plt.rcParams['font.family']='serif'
plt.rcParams['font.size']=14
plt.rcParams['mathtext.fontset']='stix'
```

```python
_=np.loadtxt('src/data/anharmonic_factor.dat')
theta_max=_[:,0]
f=_[:,1]
```

```python
if True:
    fig=plt.figure(figsize=(9,4))
    fig.subplots_adjust(bottom=0.15, left=0.15, top = 0.99, right=0.9,wspace=0.0,hspace=0.25)
    sub = fig.add_subplot(1,1,1)
    
    T=theta_max[:-1]
    T=np.append(T,np.linspace(theta_max[-1] ,np.pi,150))

    Y=[]
    for t in T:
        Y.append(anharmonic_factor(t))
    sub.plot(T,Y,linestyle='-',linewidth=2,alpha=1,c='xkcd:black',label=r"c++")
    
    sub.scatter(theta_max,f,marker='+',color='xkcd:blue')
    
    sub.set_xlabel(r'$\theta_{\rm max}$')
    sub.xaxis.set_label_coords(0.5, -0.1) 
    sub.set_ylabel(r'$ \dfrac{2\sqrt{2}}{\pi \ \theta_{\rm max}^2} \ \int_{-\theta_{\rm max}}^{\theta_{\rm max}} d\theta \ \sqrt{ \cos(\theta) - \cos(\theta_{\rm max}) }$')
    sub.yaxis.set_label_coords(-0.1,0.5) 

    
    fig.show()
```
