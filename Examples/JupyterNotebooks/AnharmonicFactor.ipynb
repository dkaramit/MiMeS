```python
# plot data alongsode the interpolation (if there are large differences there is something wrong)
if True:
    fig=plt.figure(figsize=(9,4))
    fig.subplots_adjust(bottom=0.15, left=0.15, top = 0.95, right=0.9,wspace=0.0,hspace=0.0)
    sub = fig.add_subplot(1,1,1)
    
    
    Y=[]
    T=np.linspace(0,np.pi,50)
    for t in T:
        Y.append(anharmonicFactor(t))
    sub.plot(T,Y,linestyle='-',linewidth=2,alpha=1,c='xkcd:black')
    
    
    sub.set_xlabel(r'$\theta_{\rm max}$')
    sub.xaxis.set_label_coords(0.5, -0.1) 
#     sub.set_ylabel(r'$ \dfrac{2\sqrt{2}}{\pi \ \theta_{\rm max}^2} \ \int_{-\theta_{\rm max}}^{\theta_{\rm max}} d\theta \ \sqrt{ \cos(\theta) - \cos(\theta_{\rm max}) }$')
    sub.set_ylabel(r'$ f\left( \theta_{\rm max} \right)$')
    sub.yaxis.set_label_coords(-0.1,0.5) 
    #set major ticks
    _M_xticks=[ 0,0.5,1,1.5,2,2.5,3]
    _M_yticks=[ 0.5,0.6,0.7,0.8,0.9,1 ]

    #set major ticks that will not have a label
    _M_xticks_exception=[]
    _M_yticks_exception=[]

    _m_xticks=[np.pi]
    _m_yticks=[]  
    ft=FT(_M_xticks,_M_yticks,
                 _M_xticks_exception,_M_yticks_exception,
                 _m_xticks,_m_yticks,
                 xmin=0,xmax=np.pi,ymin=0.5,ymax=1.005,xscale='linear',yscale='linear')

    ft.format_ticks(plt,sub)    
    
    sub.text(np.pi*0.99,0.46,r'$\pi$')

    
#     fig.savefig('anharmonic_factor.pdf',bbox_inches='tight')
    fig.show()
```
